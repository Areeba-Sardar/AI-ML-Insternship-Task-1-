The goal of this project is to explore and visualize the Iris dataset using Python libraries such as pandas, matplotlib, and seaborn.

This project helps understand:
- Data loading
- Data exploration
- Summary statistics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

The Iris dataset is loaded using seaborn.

```python
df = sns.load_dataset("iris")
```

# Display First 5 Rows

Using `.head()` to inspect the dataset.

```python
df.head()
```

# Dataset Shape and Columns

```python
print("Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)
```

# Dataset Information

Using `.info()` to understand:
- data types
- null values
- columns

```python
df.info()
```

# Statistical Summary

Using `.describe()` for statistics.

```python
df.describe()
```

# Scatter Plot

Relationship between sepal length and petal length.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x='sepal_length',
    y='petal_length',
    hue='species',
    data=df
)

plt.title("Sepal Length vs Petal Length")

plt.show()
```

# Histograms

Distribution of feature values.

```python
df.hist(figsize=(10,8))

plt.show()
```

# Box Plots

Used to identify outliers in the dataset.

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=df)

plt.title("Box Plot of Iris Features")

plt.show()
```

# Final Insights

- The Iris dataset contains 150 rows and 5 columns.
- Different flower species show clear differences in petal and sepal measurements.
- Scatter plots help visualize relationships between features.
- Histograms show data distribution.

- Setosa species appears clearly separated from the other species.
