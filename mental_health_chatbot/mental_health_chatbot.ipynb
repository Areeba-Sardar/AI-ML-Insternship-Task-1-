# Mental Health Support Chatbot

## Fine-Tuned AI Chatbot using DistilGPT2

## Problem Statement

The goal of this project is to build a Mental Health Support Chatbot that provides empathetic and supportive responses to users experiencing stress, anxiety, or emotional difficulties.

The chatbot is fine-tuned using the EmpatheticDialogues dataset and the DistilGPT2 language model from Hugging Face Transformers.

```python
from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    Trainer,
    TrainingArguments
)

import torch
```

## Dataset Loading

We use the EmpatheticDialogues dataset developed by Facebook AI for training empathetic conversations.

```python
dataset = load_dataset(
    "empathetic_dialogues",
    trust_remote_code=True
)

print(dataset)
```

## Dataset Exploration

Checking sample conversations from the dataset.

```python
print(dataset["train"][0])
```

## Loading DistilGPT2 Model

```python
model_name = "distilgpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token
```

## Data Preprocessing

Preparing the dialogue text for model training.

```python
def preprocess(example):
    text = "User: " + example["prompt"] + "\nBot: " + example["utterance"]

    encoding = tokenizer(
        text,
        truncation=True,
        padding="max_length",
        max_length=128
    )

    encoding["labels"] = encoding["input_ids"].copy()

    return encoding
```

```python
train_data = dataset["train"].select(range(1000))

tokenized_data = train_data.map(preprocess)
```

## Training Configuration

```python
training_args = TrainingArguments(
    output_dir="./fine_tuned_model",
    per_device_train_batch_size=2,
    num_train_epochs=1,
    logging_steps=50,
    save_steps=200,
    save_total_limit=1
)
```

## Model Training using Hugging Face Trainer API

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_data
)
```

```python
trainer.train()
```

## Saving Fine-Tuned Model

```python
model.save_pretrained("./fine_tuned_model")
tokenizer.save_pretrained("./fine_tuned_model")

print("Training Complete!")
```

## Results and Insights

The chatbot was successfully fine-tuned using DistilGPT2 on the EmpatheticDialogues dataset.

The model learned to generate supportive and emotionally aware responses. Training loss gradually decreased during training, showing improvement in conversational behavior.

A Streamlit interface was also created to interact with the chatbot in real time.
