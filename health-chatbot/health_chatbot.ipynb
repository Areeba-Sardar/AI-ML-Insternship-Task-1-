## General Health Query Chatbot

## Objective
Build a chatbot that answers general health-related questions using a Large Language Model (LLM).

```python
from transformers import pipeline
```

## Load Language Model

```python
chatbot = pipeline(
    task="text-generation",
    model="gpt2"
)
```

## Safety Filter
The chatbot blocks harmful medical-related queries.

```python
dangerous_keywords = [
    "suicide",
    "kill myself",
    "overdose",
    "self-harm",
    "harm myself"
]
```

## Predefined Health Responses
Some common health questions are answered directly for better accuracy.

```python
health_answers = {
    "sore throat": "A sore throat may be caused by viral infections, allergies, dry air, or cold weather.",
    "fever": "Fever can happen because of infections or illnesses. Drink fluids and rest well.",
    "headache": "Headaches may happen due to stress, dehydration, lack of sleep, or illness.",
    "paracetamol": "Paracetamol is generally safe when used correctly. Children should take proper doses recommended by a doctor."
}
```

## Chatbot Function
This function handles user questions and generates responses.

```python
def ask_health_question(question):

    if any(word in question.lower() for word in dangerous_keywords):
        return "Please contact a doctor or emergency service immediately."

    for keyword in health_answers:
        if keyword in question.lower():
            return health_answers[keyword]

    prompt = f"""
You are a helpful medical assistant.

Give a short and simple health answer.

Question: {question}

Answer:
"""

    response = chatbot(
        prompt,
        max_new_tokens=20,
        temperature=0.2
    )

    text = response[0]["generated_text"]

    answer = text.replace(prompt, "").strip()

    answer = answer.split(".")[0] + "."

    return answer
```

## Testing the Chatbot

```python
print(ask_health_question("What causes sore throat?"))
```

```python
print(ask_health_question("Is paracetamol safe for children?"))
```

```python
print(ask_health_question("I want to harm myself"))
```

## Results and Insights

- The chatbot successfully answers basic health-related questions.
- Prompt engineering improves response quality.
- Safety filters help block dangerous medical-related queries.
- GPT-2 was used as a free open-source language model.
